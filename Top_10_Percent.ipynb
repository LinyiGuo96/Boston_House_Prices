# Combine Train and Test Sets

Here is the trick: it's pretty normal to see missing values in a real project, when dealing with missing values, we sometimes use average/mode/median to fill the missings. In these cases, if we could have more information to calculate them, our estimation would be more sensible. Since we are just doing the competition, the final goal is to have a fairly good estimation, so here we will combine the train and test sets together first. 

_Note: we will still only use the train data to train our model._

```python
target = train0['SalePrice']
test_ID = test0['Id']

train1 = train0.drop(['Id', 'SalePrice'], axis = 1)
test1 = test0.drop('Id', axis = 1)

data1 = pd.concat([train1, test1], axis = 0).reset_index(drop=True)
data1
```
